```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import make_blobs
from sklearn.random_projection import johnson_lindenstrauss_min_dim
from sklearn.random_projection import GaussianRandomProjection, SparseRandomProjection
```

```python
X, X_labels = make_blobs(100, 10, centers=4, cluster_std=1.3)
```

Gaussian Random Project

```python
transformer = GaussianRandomProjection(3, random_state = 1)
X_new = transformer.fit_transform(X)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_new[:,0], X_new[:,1], X_new[:,2], c = X_labels)
```

```python
a = np.random.rand(1,10)
a_new = transformer.transform(a)
a_new
```

```python
M = transformer.components_
np.matmul(a,M.T)
```

Sparse Random Projection


```python
transformer = SparseRandomProjection(3, random_state = 1)
X_new = transformer.fit_transform(X)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_new[:,0], X_new[:,1], X_new[:,2], c = X_labels)
```

2d to 2d

```python
X, X_labels = make_blobs(100, 2, centers=3, cluster_std=1.3)
plt.scatter(X[:,0], X[:,1], c = X_labels)
```
